```python
from langchain_chroma import Chroma
from langchain_community.embeddings.sentence_transformer import SentenceTransformerEmbeddings
from langchain_community.document_loaders import TextLoader

loader = TextLoader("releve_de_notes.txt")
documents = loader.load()
# Texte directement fourni
text = """
President Biden delivered a speech during the State of the Union address, 
where he discussed various topics, including new appointments and the administration's focus on justice and equality.
"""

# Création de la fonction d'embedding utilisant Sentence Transformers
embedding_function = SentenceTransformerEmbeddings(model_name="all-MiniLM-L6-v2")

# Initialisation de Chroma
db = Chroma()

# Encodage du texte en embedding et chargement dans Chroma
# Note: 'from_documents' et 'similarity_search' sont supposés être des méthodes de la classe Chroma
# Ces méthodes doivent être adaptées en fonction de l'API réelle disponible pour Chroma.
document_embedding = embedding_function.encode([text])
db.from_documents([{'text': text, 'embedding': document_embedding}], embedding_function)

# Recherche de similarité dans Chroma
query = "What did the president say about justice and equality?"
similar_docs = db.similarity_search(query)

# Affichage des résultats
if similar_docs:
    print("Résultat de la recherche de similarité :", similar_docs[0]['text'])
else:
    print("Aucun document similaire trouvé.")
```
